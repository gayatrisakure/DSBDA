```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
req=requests.get('https://www.ebay.com/p/18034763686?iid=166029124414&rt=nc')
req
```

```python
req.text
```

```python
req.content
```

```python
soup=bs(req.text)
soup
```

```python
rating=soup.find('span',{'class':'review--start--rating'}).get_text()
print(rating)
```

```python
comments=soup.find_all('h4',{'class':'review--title'})
for comment in comments:
    print(comment.get_text()+"\n\n")
```

```python
reviews=soup.find_all('p',{'class':'review--content'})
for review in reviews:
    print(review.get_text()+"\n\n")
```

```python
cust=soup.find_all('a',{'class':'review--author'})
for c in cust:
    print(c.get_text()+"\n\n")
```

```python
data={
    "name":cust,
    "review":reviews,
    "rating":rating,
    "comment":comments
}
```

```python
pd.set_option('display.max_colwidth',None)
df=pd.DataFrame(data)
df
```

```python
df.to_csv('29.csv')
```
